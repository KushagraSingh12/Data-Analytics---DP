```python
a.add(2)
a.add(4)
a.add(3)
a.add(2)
a.add(8)
a.add(10)
print(a)
```

```python
a.remove(4) #it will throw an error if item does not exist in the collection
print(a)
```

```python
a.discard(2) #it will not throw an error if item does not exist in the collection
print(a)
```

```python
a ={2,3,4,5}
b = {4,5,6,7}
print(a)
print(b)
```

```python
#union
print(a.union(b))
print(a|b) # this | operator is also known as union
```

```python
#intersection - all items that are common in both
print(a.intersection(b))
print(a&b) # & this operator is known as intersection
```

```python
#diffrence 
print(a.difference(b))
print(b-a)
```

```python
#symmetric diffrence 
print(a.symmetric_difference(b))
print(a^b) #^this operator used for symmetric diffrence
```
