now lets see manipulation.


```python
#add new key value pair 
info['college'] = 'LU'
print(info)
```

```python
#update key value pair 
info['college'] = 'LPC'
print(info)
```

```python
#delete key value pair
info.pop('college')
print(info)
```

```python
print(info.keys())
print(info.values())
print(info.items())
```

```python
#k here is a pointer variable
#agar single pointer variable is given it gets assigned to keys

for K in info:
    print(K)
```

```python
a = 10
b = 20
print("sum is :", a+b)
#use f string to reduce comma 
print(f"sum of {a} and {b} is {a+b}")
```

```python

for K,v in info.items():
    print(f'{K}: {v}')
```
