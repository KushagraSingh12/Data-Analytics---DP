```python
#3.Non-returning function

def avg(a,b,c):
    print((a+b+c)/3)
    
average = avg(10,20,30)
print(average)
    
```

```python
#Returning function

def avg(a,b,c):
    return((a+b+c)/3)
    
average = avg(10,20,30)
print(average)
    
```
