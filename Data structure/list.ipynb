there are two types of indexing
1. postive - which starts from zero
2. negative - which strats from -1
use negative indexing when print list from the last 

```python
print(apps[0])
print(apps[7])
```

```python
print(apps[-2])
print(apps[-6])
```

SLICING
USE OF COLLEN (:) HERE 
SLICING- WHEN U WANT TO A SUB LIST FROM A LIST .

```python
print(apps[1:8:2])
print(apps[:5])
print(apps[2:])
print(apps[:])
```

flow of indexing is always left to right

```python
print(apps[-4:])
```

LIST METHODS 
for manupulating the list.


append() -to add item at the end of the list 
insert() -to add item at any specific postion in the list
pop()    -to delete item using index option
remove() -to delete items using item name 
copy()   -to make duplicate of the list
sort()   -to sort list in a particular order
extend() -to combine two list into a single list
clear()  -to delete all the items from the list but mot the list

```python
apps.append('watsapp')
print(apps)
```

```python
apps.insert(2,'netflix')
print(apps)
```

```python
apps.pop(0)
print(apps)
```

```python
apps.remove('map')
print(apps)
```

```python
a = apps.copy()
print(a)
```

```python
a.clear()
print(a)
```

```python
apps.sort()
print(apps)
```

```python
apps2 =[1,2,3,4,5]
apps.extend(apps2)
print(apps)
```
